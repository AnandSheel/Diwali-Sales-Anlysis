# Exploratory Data Analysis

## Gender

```python
df.columns
```

```python
ax = sns.countplot( x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
```

```python
df.groupby(['Gender'], as_index=False).agg({'Amount': 'sum'}).sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False).agg({'Amount': 'sum'}).sort_values(by='Amount', ascending=False)
sns.barplot( x= 'Gender', y= 'Amount', data= sales_gen)
```

*From above graphs we can see that most of the buyers are females and evn the purchacing power of females are greater than males*

## Age

```python
df.columns
```

```python
ax = sns.countplot(data = df, x= 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group

sales_age= df.groupby(['Age Group'], as_index=False).agg({'Amount': 'sum'}).sort_values(by='Amount', ascending=False)

sns.barplot( x = 'Age Group', y = 'Amount', data = sales_age)
```

*Frome above graphs we can see that most of the buyers are of age group between 26-35 yrs Female*

## State

```python
df.columns
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False).agg({'Orders': 'sum'}).sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(19, 10)})
sns.barplot( data = sales_state, x= 'State', y= 'Orders' )
```

```python
sales_state = df.groupby(['State'], as_index=False).agg({'Amount': 'sum'}).sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot( data = sales_state, x= 'State', y= 'Amount' )
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

## Marital Status

```python
df.columns
```

```python
ax = sns.countplot( data = df, x = 'Marital_Status')

sns.set( rc = {'figure.figsize':(6,3)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False).agg({'Amount': 'sum'}).sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot( data = sales_state, x= 'Marital_Status', y= 'Amount', hue = 'Gender' )
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

## Occupation

```python
sns.set(rc ={'figure.figsize': (19,10)})
ax = sns.countplot(data = df, x= 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False).agg({'Amount': 'sum'}).sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot( data = sales_state, x= 'Occupation', y= 'Amount' )
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

## Product Category

```python
sns.set(rc={'figure.figsize':(25,10)})
ax = sns.countplot(data= df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False).agg({'Amount': 'sum'}).sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,10)})
sns.barplot( data = sales_state, x= 'Product_Category', y= 'Amount' )
```

*From the above graphs we can see that most of the sold products are from Food,Cothing and Electronics category.*

```python
sales_state = df.groupby(['Product_ID'], as_index=False).agg({'Orders': 'sum'}).sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,10)})
sns.barplot( data = sales_state, x= 'Product_ID', y= 'Orders' )
```

```python
# top 10 most sold products 

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby(['Product_ID'], as_index=False).agg({'Orders': 'sum'}).sort_values(by='Orders', ascending=False).head(10).plot(kind='bar')
```
